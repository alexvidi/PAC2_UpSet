```python
from upsetplot import from_memberships


movies_by_genre = from_memberships(movies.Genre.str.split(','), data=movies)
movies_by_genre


```

```python
from upsetplot import UpSet

UpSet(movies_by_genre, min_subset_size=15, show_counts=True).plot()
```
